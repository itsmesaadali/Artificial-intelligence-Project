```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris, load_wine, make_classification
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, f1_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import warnings
warnings.filterwarnings('ignore')

# Set random seed for reproducibility
np.random.seed(42)
```

```python
# Generate a custom binary classification dataset
print("=" * 60)
print("DATASET 1: CUSTOM BINARY CLASSIFICATION")
print("=" * 60)

X1, y1 = make_classification(
    n_samples=1000,
    n_features=4,
    n_informative=2,
    n_redundant=1,
    n_repeated=1,
    n_classes=2,
    n_clusters_per_class=1,
    random_state=42
)

# Split the dataset
X1_train, X1_test, y1_train, y1_test = train_test_split(
    X1, y1, test_size=0.3, random_state=42
)

# Create and train Gaussian Naive Bayes model
gnb1 = GaussianNB()
gnb1.fit(X1_train, y1_train)

# Make predictions
y1_pred = gnb1.predict(X1_test)

# Evaluate the model
accuracy1 = accuracy_score(y1_test, y1_pred)
f1_1 = f1_score(y1_test, y1_pred)

print(f"Accuracy: {accuracy1:.4f}")
print(f"F1 Score: {f1_1:.4f}")
print("\nClassification Report:")
print(classification_report(y1_test, y1_pred))

# Plot confusion matrix
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
cm1 = confusion_matrix(y1_test, y1_pred)
sns.heatmap(cm1, annot=True, fmt='d', cmap='Blues')
plt.title('Dataset 1: Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
```

```python
# Load Iris dataset (Multiclass)
print("\n" + "=" * 60)
print("DATASET 2: IRIS DATASET (MULTICLASS)")
print("=" * 60)

iris = load_iris()
X2, y2 = iris.data, iris.target

# Split the dataset
X2_train, X2_test, y2_train, y2_test = train_test_split(
    X2, y2, test_size=0.3, random_state=42
)

# Create and train Gaussian Naive Bayes model
gnb2 = GaussianNB()
gnb2.fit(X2_train, y2_train)

# Make predictions
y2_pred = gnb2.predict(X2_test)

# Evaluate the model
accuracy2 = accuracy_score(y2_test, y2_pred)
f1_2 = f1_score(y2_test, y2_pred, average='weighted')

print(f"Accuracy: {accuracy2:.4f}")
print(f"F1 Score (weighted): {f1_2:.4f}")
print("\nClassification Report:")
print(classification_report(y2_test, y2_pred, target_names=iris.target_names))

# Plot confusion matrix
plt.subplot(1, 3, 2)
cm2 = confusion_matrix(y2_test, y2_pred)
sns.heatmap(cm2, annot=True, fmt='d', cmap='Greens',
            xticklabels=iris.target_names,
            yticklabels=iris.target_names)
plt.title('Dataset 2: Iris Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
```

```python
# Load Wine dataset (Multiclass)
print("\n" + "=" * 60)
print("DATASET 3: WINE DATASET (MULTICLASS)")
print("=" * 60)

wine = load_wine()
X3, y3 = wine.data, wine.target

# Split the dataset
X3_train, X3_test, y3_train, y3_test = train_test_split(
    X3, y3, test_size=0.3, random_state=42
)

# Create and train Gaussian Naive Bayes model
gnb3 = GaussianNB()
gnb3.fit(X3_train, y3_train)

# Make predictions
y3_pred = gnb3.predict(X3_test)

# Evaluate the model
accuracy3 = accuracy_score(y3_test, y3_pred)
f1_3 = f1_score(y3_test, y3_pred, average='weighted')

print(f"Accuracy: {accuracy3:.4f}")
print(f"F1 Score (weighted): {f1_3:.4f}")
print("\nClassification Report:")
print(classification_report(y3_test, y3_pred, target_names=wine.target_names))

# Plot confusion matrix
plt.subplot(1, 3, 3)
cm3 = confusion_matrix(y3_test, y3_pred)
sns.heatmap(cm3, annot=True, fmt='d', cmap='Reds',
            xticklabels=wine.target_names,
            yticklabels=wine.target_names)
plt.title('Dataset 3: Wine Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')

plt.tight_layout()
plt.show()
```

```python
# Compare different Naive Bayes variants and parameters
print("\n" + "=" * 70)
print("COMPARISON OF DIFFERENT NAIVE BAYES VARIANTS AND PARAMETERS")
print("=" * 70)

# Use Iris dataset for comparison
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Different Naive Bayes classifiers
classifiers = {
    'Gaussian Naive Bayes': GaussianNB(),
    'Gaussian NB (var_smoothing=1e-9)': GaussianNB(var_smoothing=1e-9),
    'Gaussian NB (var_smoothing=1e-5)': GaussianNB(var_smoothing=1e-5),
}

# Apply different preprocessing
preprocessors = {
    'No Scaling': None,
    'Standard Scaling': StandardScaler(),
    'MinMax Scaling': MinMaxScaler()
}

results = []

for preproc_name, preprocessor in preprocessors.items():
    print(f"\n--- {preproc_name} ---")

    # Apply preprocessing if specified
    if preprocessor is not None:
        X_train_processed = preprocessor.fit_transform(X_train)
        X_test_processed = preprocessor.transform(X_test)
    else:
        X_train_processed = X_train.copy()
        X_test_processed = X_test.copy()

    for clf_name, classifier in classifiers.items():
        # Train classifier
        classifier.fit(X_train_processed, y_train)

        # Make predictions
        y_pred = classifier.predict(X_test_processed)

        # Calculate metrics
        accuracy = accuracy_score(y_test, y_pred)
        f1 = f1_score(y_test, y_pred, average='weighted')

        results.append({
            'Preprocessor': preproc_name,
            'Classifier': clf_name,
            'Accuracy': accuracy,
            'F1 Score': f1
        })

        print(f"{clf_name}: Accuracy = {accuracy:.4f}, F1 = {f1:.4f}")

# Create results dataframe
results_df = pd.DataFrame(results)
print("\n" + "=" * 70)
print("SUMMARY OF RESULTS")
print("=" * 70)
print(results_df.to_string(index=False))
```

```python
# Advanced implementation with feature importance analysis
print("\n" + "=" * 70)
print("FEATURE ANALYSIS WITH NAIVE BAYES")
print("=" * 70)

# Use Wine dataset for feature analysis
wine = load_wine()
X, y = wine.data, wine.target
feature_names = wine.feature_names

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Gaussian Naive Bayes
gnb = GaussianNB()
gnb.fit(X_train, y_train)

# Make predictions
y_pred = gnb.predict(X_test)

# Calculate feature importance (based on standard deviations)
feature_importance = np.std(gnb.theta_, axis=0)
feature_importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': feature_importance
}).sort_values('Importance', ascending=False)

print("Feature Importance (based on class-wise feature standard deviations):")
print(feature_importance_df.to_string(index=False))

# Plot feature importance
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.barplot(data=feature_importance_df, x='Importance', y='Feature', palette='viridis')
plt.title('Feature Importance in Naive Bayes Model')
plt.xlabel('Importance (Std Dev of Class Means)')

# Plot class distributions for top 2 features
plt.subplot(1, 2, 2)
top_features = feature_importance_df.head(2)['Feature'].values
feature1_idx = list(feature_names).index(top_features[0])
feature2_idx = list(feature_names).index(top_features[1])

plt.scatter(X_train[:, feature1_idx], X_train[:, feature2_idx], c=y_train,
            cmap='viridis', alpha=0.7, edgecolors='w', s=50)
plt.xlabel(top_features[0])
plt.ylabel(top_features[1])
plt.title(f'Class Distribution: {top_features[0]} vs {top_features[1]}')
plt.colorbar(label='Class')

plt.tight_layout()
plt.show()

# Final model evaluation
print("\n" + "=" * 70)
print("FINAL COMPREHENSIVE MODEL EVALUATION")
print("=" * 70)

accuracy = accuracy_score(y_test, y_pred)
f1_weighted = f1_score(y_test, y_pred, average='weighted')
f1_macro = f1_score(y_test, y_pred, average='macro')

print(f"Accuracy: {accuracy:.4f}")
print(f"F1 Score (Weighted): {f1_weighted:.4f}")
print(f"F1 Score (Macro): {f1_macro:.4f}")

print("\nDetailed Classification Report:")
print(classification_report(y_test, y_pred, target_names=wine.target_names))

# Final confusion matrix
plt.figure(figsize=(8, 6))
cm_final = confusion_matrix(y_test, y_pred)
sns.heatmap(cm_final, annot=True, fmt='d', cmap='coolwarm',
            xticklabels=wine.target_names,
            yticklabels=wine.target_names)
plt.title('Final Model: Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
# Performance summary across all experiments
print("\n" + "=" * 70)
print("OVERALL PERFORMANCE SUMMARY")
print("=" * 70)

summary_data = {
    'Dataset': ['Custom Binary', 'Iris', 'Wine'],
    'Classes': [2, 3, 3],
    'Samples': [X1.shape[0], X2.shape[0], X3.shape[0]],
    'Features': [X1.shape[1], X2.shape[1], X3.shape[1]],
    'Accuracy': [accuracy1, accuracy2, accuracy3],
    'F1 Score': [f1_1, f1_2, f1_3]
}

summary_df = pd.DataFrame(summary_data)
print(summary_df.to_string(index=False))

# Plot performance comparison
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
x_pos = np.arange(len(summary_df))
width = 0.35

plt.bar(x_pos - width/2, summary_df['Accuracy'], width, label='Accuracy', alpha=0.7)
plt.bar(x_pos + width/2, summary_df['F1 Score'], width, label='F1 Score', alpha=0.7)

plt.xlabel('Dataset')
plt.ylabel('Score')
plt.title('Performance Comparison Across Datasets')
plt.xticks(x_pos, summary_df['Dataset'])
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
# Best performing configuration from parameter tuning
best_result = results_df.loc[results_df['Accuracy'].idxmax()]
best_config = f"{best_result['Preprocessor']}\n{best_result['Classifier']}"

configs = [f"Config {i+1}" for i in range(len(results_df))]
accuracies = results_df['Accuracy']

plt.bar(configs, accuracies, color='lightcoral', alpha=0.7)
plt.axhline(y=best_result['Accuracy'], color='red', linestyle='--',
            label=f'Best: {best_result["Accuracy"]:.4f}')
plt.xlabel('Configuration')
plt.ylabel('Accuracy')
plt.title('Accuracy Across Different Configurations')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\nBest Configuration:")
print(f"Preprocessor: {best_result['Preprocessor']}")
print(f"Classifier: {best_result['Classifier']}")
print(f"Accuracy: {best_result['Accuracy']:.4f}")
print(f"F1 Score: {best_result['F1 Score']:.4f}")
```
